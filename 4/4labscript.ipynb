# [dbo].[AddAccount]

```sql
/****** Object:  StoredProcedure [dbo].[AddAccount]    Script Date: 17.03.2022 22:42:49 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[AddAccount]
    @login NVARCHAR(50), 
    @password NVARCHAR(50), 
    @result NVARCHAR(250) OUTPUT
AS
BEGIN
    SET NOCOUNT ON
    BEGIN TRY
        INSERT INTO Accounts 
        VALUES(@login, HASHBYTES('SHA2_512', @password))

        SET @result='Success'

    END TRY
    BEGIN CATCH
        SET @result=ERROR_MESSAGE() 
    END CATCH

END
GO
```

# [dbo].[UserLogin]

```sql
/****** Object:  StoredProcedure [dbo].[UserLogin]    Script Date: 17.03.2022 22:42:49 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UserLogin]
    @login NVARCHAR(50),
    @password NVARCHAR(50),
    @result NVARCHAR(250)='' OUTPUT
AS
BEGIN
    SET NOCOUNT ON
    DECLARE @Id INT

    IF EXISTS (SELECT TOP 1 id FROM [dbo].[Accounts] WHERE [login]=@login)
    BEGIN
       SET @Id=(SELECT id FROM [dbo].[Accounts] WHERE [login]=@login AND PasswordHash=HASHBYTES('SHA2_512', @password))
       IF(@Id IS NULL)
           SET @result='Incorrect password'
       ELSE 
           SET @result='User successfully logged in'
    END
    ELSE
       SET @result='Invalid login'
END


GO
```
